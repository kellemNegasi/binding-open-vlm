```python
import numpy as np
from sklearn.model_selection import train_test_split

embeddings_path = "output/probing_global/scene_description_balanced_2d_limited_qwen2.5-VL-7B-Instruct/global_embeddings.npz"
test_split = 0.2
seed = 0

with np.load(embeddings_path) as z:
    triplet = z["triplet_count_per_sample"]
    implied = z["y_implied_absent"].astype(bool)

idx = np.arange(len(triplet))
idx_train, idx_test = train_test_split(
    idx,
    test_size=test_split,
    random_state=seed,
    stratify=triplet if len(np.unique(triplet)) > 1 else None,
)

triplet_test = triplet[idx_test]
implied_test = implied[idx_test]

print("triplet\t\t#samples\t#implied_absent_pairs\tmean_pairs_per_sample")
for t in sorted(np.unique(triplet_test)):
    m = triplet_test == t
    n_samples = int(m.sum())
    n_pairs = int(implied_test[m].sum())
    mean_pairs = float(implied_test[m].sum(axis=1).mean()) if n_samples else 0.0
    print(f"{t}\t\t{n_samples}\t\t\t{n_pairs}\t\t\t{mean_pairs:.2f}")
```
